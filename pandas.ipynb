```python
#handle missing values using mean,mode
#real life use=cleaning  raw data
import pandas as pd
data= {
    "sal":[20000,40000,75000],
    "package":[3.5,4.5,8],
    "name":["shravani","riya","renu"],
    "Marks":[85,45,90,70],
    "Age": [25, 30, 25, 22, 30],
    "City": ["Amt", "Pune", "Mumbai", "banglore", "Pune"]
}
df[['marks', 'age']] = df[['marks', 'age']].fillna(df[['marks', 'age']].mean())
df=pd.DataFrame(data)
df["Marks"].mean()
df["Marks"].mode()
```
